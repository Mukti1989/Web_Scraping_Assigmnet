# Q-9.Write a python program to scrape mentioned details from dineout.co.in :
    i)-Restaurant name
    ii)-Cuisine
    iii)-Location
    iv)-Ratings
    v)-Image URL

```python
page = requests.get('https://www.dineout.co.in/delhi-restaurants/welcome-back')
page
```

```python
soup =  BeautifulSoup(page.content)
soup
```

```python
rest_name = []

for i in soup.find_all('div', class_="restnt-info cursor"):
    rest_name.append(i.text.replace('\n', ""))
rest_name
```

```python
Cuisine = []

for i in soup.find_all('div', class_="detail-info"):
    Cuisine.append(i.text.replace('\n', ""))
Cuisine
```

```python
location = []

for i in soup.find_all('div', class_="restnt-loc ellipsis"):
    location.append(i.text.replace('\n', ""))
location
```

```python
rating = []

for i in soup.find_all('div', class_="restnt-rating rating-4"):
    rating.append(i.text.replace('\n', ""))
rating
```

```python
image_url = []

for i in soup.find_all('img', class_="no-img"):
    image_url.append(i['data-src'])
image_url
```
