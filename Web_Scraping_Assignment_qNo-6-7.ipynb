# Q-6.Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:

# A-Top 10 ODI teams in women’s cricket along with the records for matches, points and rating.

```python
page = requests.get('https://www.icc-cricket.com/rankings/womens/team-rankings/odi')
page
```

```python
soup =  BeautifulSoup(page.content)
soup
```

```python
team1 = []

for i in soup.find_all('span', class_="u-hide-phablet"):
    team1.append(i.text.replace('\n', ''))
team1[0:10]
```

```python
matches = []

for i in soup.find_all('td', class_="rankings-block__banner--matches"):
    matches.append(i.text.replace('\n', ''))
matches[0:10]
```

```python
matches1 = []

for i in soup.find_all('td', class_="table-body__cell u-center-text"):
    matches1.append(i.text.replace('\n', ''))
matches1[0:20:2]
```

```python
Matches = (matches+matches1)
Matches[0:21:3]
```

```python
point1 = []

for i in soup.find_all('td', class_="table-body__cell u-center-text"):
    point1.append(i.text.replace('\n', ''))
point1[0:20:2]
```

```python
rating = []

for i in soup.find_all('td', class_="rankings-block__banner--rating u-text-right"):
    rating.append(i.text.replace('\n', ''))
rating
```

# Q-7.Write a python program to scrape details of all the posts from coreyms.com. Scrape the heading, date, content and the code for the video from the link for the youtube video from the post

```python
page3 = requests.get('https://coreyms.com./')
page3
```

```python
soup = BeautifulSoup(page3.content)
soup
```

```python
heading = []

for i in soup.find_all('h2', class_="entry-title"):
    heading.append(i.text.replace('\n', ''))
heading
```

```python
Date = []

for i in soup.find_all('time', class_="entry-time"):
    Date.append(i.text.replace('\n', ''))
Date
```

```python
Content = []

for i in soup.find_all('div', class_="entry-content"):
    Content.append(i.text.replace('\n', ''))
Content
```

```python
youtube = []

for i in soup.find_all('div', class_="ytp-title"):
    youtube.append(i.text.replace('\n', ''))
youtube
```
